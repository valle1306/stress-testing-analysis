```python
import pandas as pd
import numpy as np
```

```python
all_participants = pd.read_csv("C:/Users/lpnhu/Downloads/Stress_Testing_Analysis/data/cleaned_master_df.csv")
all_participants.head()
```

```python
stress = pd.read_csv('stress_labels_all_participants.csv')
stress
```

```python
# Convert to datetime
all_participants["date"] = pd.to_datetime(all_participants["timestamp_iso"]).dt.date
stress["date"] = pd.to_datetime(stress["date"]).dt.date

# Merge on participant_id and date
merged_df = pd.merge(all_participants, stress, on=["participant_id", "date"], how="inner")

# Preview
print(merged_df.shape)
merged_df.head()
```

```python
merged_df.to_csv("C:/Users/lpnhu/Downloads/Stress_Testing_Analysis/data/merged_all_participants_with_stress.csv", index=False)
```
