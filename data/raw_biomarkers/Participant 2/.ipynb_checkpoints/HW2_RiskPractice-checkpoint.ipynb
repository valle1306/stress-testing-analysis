### Plot daily returns

```python
returns = {ticker: tickers_df['Close'].apply(log).diff()[1:]}
train = returns[ticker] * 100
train2 = train * train

plt.figure(figsize=(18,6))
plt.plot(train)
plt.title(ticker, fontsize=25)
plt.plot([train.index[0], train.index[-1]], [0,0], '--',color='red')
plt.ylabel('daily return', fontsize=25) 
plt.show()
```

### MA: fit interval moving average value at risk

```python
history_n = 40  # length of window for calculating historical volatility
MAVaR = [-1.645 * np.sqrt(np.mean(train2[i:i+history_n])) for i,_ in enumerate(train2[:-history_n])]

MAVaR_df = pd.DataFrame(MAVaR[:], index=pd.DataFrame(train).index[40:], columns=['VaR'])
plt.figure(figsize=(18,8))
start_date = '2018-01-01'
plt.plot(MAVaR_df[start_date:]['VaR'],linewidth=3)
plt.plot(pd.DataFrame(train[start_date:]), 'ro')
plt.plot(pd.DataFrame([0,0],index=MAVaR_df[start_date:].index[[0,-1]]), 'g-',linewidth=3)
plt.title("moving average", fontsize=25)
plt.ylabel('daily return', fontsize=25)
plt.show()
```

### EMWA

For _vol_df['2018-01-01':]['EMWA']:
* the vol_df['vol'] derived from volatile history, which tracks expoenentially weighted MA values of train
* so the full thing contains the computed EMWA values for the dates starting from 2018-01-01

For train['2018-01-01']:
* train stores the daily returns values
* The full code retrieves the return values from 2018-01-01 onwards

```python
#Calculate the 95% VaR over the same time period using EWMA(lambda = .94)
#VaR cutoff is computed using the data up to the previous day's close, so it does not include today's
## return when comparing today's VaR level

train2 = train * train

wt = .94
vol = train2[0]
vol_history = [0,vol]  # move everything forward one day
for v in train2[1:]:
    vol = wt * vol + (1 - wt) * v
    vol_history.append(vol)
    
vol_df = pd.DataFrame(vol_history[:-1], index=train.index, columns=['vol'])
vol_df['EWMA'] = [-1.64 * math.sqrt(x) for x in vol_df['vol']]

plt.figure(figsize=(18,8))
plt.plot(vol_df['2018-01-01':]['EWMA'], linewidth=3)
plt.plot(pd.DataFrame(train['2018-01-01':]), 'ro')
plt.plot(pd.DataFrame([0,0],index=MAVaR_df[start_date:].index[[0,-1]]), 'g-',linewidth=3)
plt.title('exponential weighted moving average', fontsize=25)
plt.ylabel('daily return', fontsize=25)
plt.show()
```

### For Feb 3, what is the 95% VaR in dollars if you hold 10 shares of Amazon? 

```python
import numpy as np
import pandas as pd
import yfinance as yf

# Step 1: Download Amazon (AMZN) stock data
ticker = "AMZN"
start_date = "2023-01-01"
end_date = "2024-02-03"
tickers_df = yf.download(ticker, start=start_date, end=end_date)

# Ensure the index is datetime
tickers_df.index = pd.to_datetime(tickers_df.index)

# Step 2: Compute daily log returns
tickers_df['Returns'] = np.log(tickers_df['Close'] / tickers_df['Close'].shift(1))

# Step 3: Compute EWMA standard deviation
lambda_ = 0.94  # Smoothing factor for EWMA
ewma_vol = tickers_df["Returns"].ewm(span=(2/(1-lambda_)), adjust=False).std()

# Step 4: Compute 95% VaR using the last available price
z_score = 1.64
tickers_df["VaR_95"] = -z_score * ewma_vol * tickers_df["Close"]

# Step 5: Compute VaR for 10 shares on Feb 3
date_to_check = "2024-02-02"
if date_to_check in tickers_df.index:
    var_95_feb3 = tickers_df.loc[date_to_check, "VaR_95"] * 10
else:
    closest_date = tickers_df.index[tickers_df.index.get_loc(date_to_check, method="pad")]
    var_95_feb3 = tickers_df.loc[closest_date, "VaR_95"] * 10
    print(f"Using closest available date: {closest_date}")

print(f"95% VaR on {date_to_check} for 10 shares of Amazon: ${var_95_feb3:.2f}")
```

```python
exceptions = (tickers_df["Returns"] < tickers_df["VaR_95"]).sum()

print(f"Number of exceptions (violations of 95% VaR) from {start_date} to {end_date}: {exceptions}")
```
