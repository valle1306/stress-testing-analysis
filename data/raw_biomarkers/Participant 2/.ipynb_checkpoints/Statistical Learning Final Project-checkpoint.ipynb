# CLIP with FER 2013 Dataset

The FER 2013 dataset contains facial images at relatively low resolutions, we should try to use *VIT- B/16* OR *ViT-B/32*. These models of CLIP strikes a good balance between performance and computation requirements.

The transformer based architecture can better capture nuanced facial features for emotion classification

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/16", device=device)
```

### Pre-processing

```python
def preprocess_data(df, has_labels=True):
    images = []
    labels = []
    for _, row in df.iterrows():
        # Convert space-separated pixel string to a NumPy array
        pixels = np.array(row["pixels"].split(), dtype=np.uint8).reshape(48, 48)
        image = Image.fromarray(pixels)  # Convert NumPy array to PIL image
        images.append(image)
        
        if has_labels:
            label = int(row["emotion"])  # Convert the label to an integer
            labels.append(label)
    
    if has_labels:
        return images, labels
    return images
```

```python
# Preprocess train dataset (has labels)
train_images, train_labels = preprocess_data(train, has_labels=True)

# Preprocess test dataset (no labels)
test_images = preprocess_data(test, has_labels=False)

print(f"Number of train images: {len(train_images)}")
print(f"Number of test images: {len(test_images)}")
```

```python
# Visualize a train image and its label
plt.imshow(train_images[0], cmap="gray")
plt.title(f"Train Label: {train_labels[0]}")
plt.show()

# Visualize a test image (no label)
plt.imshow(test_images[0], cmap="gray")
plt.title("Test Image")
plt.show()
```

```python
class FER2013Dataset(Dataset):
    def __init__(self, images, labels=None, transform=None):  
        self.images = images
        self.labels = labels
        self.transform = transform  # Assign the transform parameter

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        # Get the image
        image = self.images[idx]
        
        # Apply transformations (e.g., preprocess from CLIP)
        if self.transform:
            image = self.transform(image)
        
        if self.labels is not None:
            return image, self.labels[idx]
        return image

    
# Pass preprocess as the transform for the dataset
train_dataset = FER2013Dataset(train_images, train_labels, transform=preprocess)
test_dataset = FER2013Dataset(test_images, transform=preprocess)
```

### Using CLIP models for encode images and text prompts

```python
# Define emotion labels as text prompts
emotion_labels = [
    "a person looking angry", 
    "a person looking disgusted", 
    "a person looking fearful", 
    "a person looking happy", 
    "a person looking sad", 
    "a person looking surprised", 
    "a person looking neutral"
]

# Tokenize the text prompts
text_inputs = torch.cat([clip.tokenize(label) for label in emotion_labels]).to(device)
```

```python
# Encode images and text for a single batch
with torch.no_grad():
    # Encode the text prompts
    text_features = model.encode_text(text_inputs)
    
    # Encode the batch of images
    image_features = model.encode_image(images.to(device))

# Normalize the features
text_features /= text_features.norm(dim=-1, keepdim=True)
image_features /= image_features.norm(dim=-1, keepdim=True)
```

```python
# Check unique label values and their range
print(f"Unique labels in dataset: {set(labels.tolist())}")
print(f"Emotion labels length: {len(emotion_labels)}")
```

```python
# Compute similarity between images and text
similarity = image_features @ text_features.T  # Dot product

# Get the top predicted label for each image
values, indices = similarity.topk(1, dim=1)  # Get top-1 prediction
predictions = [emotion_labels[idx] for idx in indices.squeeze().tolist()]

# Calculate the number of samples to iterate over
num_samples = min(len(images), len(predictions), len(labels))

# Iterate over the samples
for i in range(num_samples):
    print(f"Image {i + 1}:")
    print(f"  Predicted: {predictions[i]}")
    print(f"  Ground Truth: {emotion_labels[labels[i]]}\n")
```

```python
### Evaluate
correct = 0
total = 0
for i in range(num_samples):
    if predictions[i] == emotion_labels[labels[i]]:
        correct += 1
    total += 1
accuracy = correct / total
print(f"Accuracy: {accuracy:.2%}")
```
