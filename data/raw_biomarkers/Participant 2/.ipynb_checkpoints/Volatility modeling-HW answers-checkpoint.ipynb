# EWMA 95% VaR for Amazon

```python
tickers=['BTC']
ticker = tickers[0]
tickers_df = yf.download(tickers, 
                      start='2015-01-01', 
                      end='2025-02-01', # set to next day to get price to 1/31
                      progress=False, auto_adjust=True)
tickers_df.head()
```

```python
returns = tickers_df['Close'].apply(log).diff()[1:]
train = returns * 100
train2 = train * train
train[:5], train2[:5]
```

```python
wt = .94
vol = train2[0]
vol_history = [0,vol]  # move everything forward one day
for v in train2[1:]:
    vol = wt * vol + (1 - wt) * v
    vol_history.append(vol)
    
EWMA_df = pd.DataFrame(vol_history[:-1], index=train.index, columns=['vol'])
EWMA_df['VaR'] = [-1.64 * math.sqrt(x) for x in EWMA_df['vol']]
EWMA_df['returns'] = train

plt.figure(figsize=(18,6))
plt.plot(EWMA_df['2018-01-01':]['VaR'])
plt.plot(EWMA_df['2018-01-01':]['returns'], 'ro')
plt.title('exponential weighted moving average', fontsize=25)
plt.ylabel('daily return', fontsize=25)
plt.show()

EWMA = EWMA_df[EWMA_df.index >= '2021-01-01']
EWMA.loc[:,'exception'] = EWMA['returns'] < EWMA['VaR']
exception_count = EWMA['exception'].sum()
day_count = EWMA.shape[0]

print(ticker, ': number of exceptions = ', exception_count, ' out of ', day_count, 'days')
print(f'expected number of exceptions = {day_count * 0.05}')
```

# 3. Count the number of exceptions and backtest the VaR estimate to determine whether this is acceptable

```python
# count the number of exceptions
EWMA_df['exception'] = EWMA_df['returns'] < EWMA_df['VaR']
exception_count = EWMA_df['exception'].sum()
```

```python
# Kupiec proportion of failures test


# Parameters
p = 0.05  # 95% confidence level
n = len(EWMA_df['exception'])  # Total observations
x = EWMA_df['exception'].sum()  # Number of exceptions

# Kupiec POF Test statistic
LR_pof = -2 * np.log(((1 - p)**(n - x) * p**x) /
                     ((1 - (x / n))**(n - x) * (x / n)**x))

# P-value
p_value = 1 - chi2.cdf(LR_pof, df=1)

# Display the result
print(f"Kupiec LR statistic: {LR_pof}")
print(f"P-value: {p_value}")
```

Since p value > 0.05, the model passes the test, indicating VaR estimate is acceptable

```python
# Identify consecutive exception periods
EWMA_df['exception_shift'] = EWMA_df['exception'].shift(1)
EWMA_df['exception_block'] = (EWMA_df['exception'] != EWMA_df['exception_shift']).cumsum()

# Filter for exception blocks
exception_blocks = EWMA_df[EWMA_df['exception'] == 1].groupby('exception_block').size()

# Find the largest block
largest_block = exception_blocks.max()

print(f"Largest block of consecutive exceptions: {largest_block}")
```

# 4. Lambda

```python
# Function to calculate EWMA, VaR, and perform Kupiec Test
def ewma_var_backtest(returns, lambda_value, confidence_level=0.95):
    # EWMA volatility calculation
    vol = returns.var()  # Initial variance
    vol_history = [vol]
    
    for r in returns:
        vol = lambda_value * vol + (1 - lambda_value) * r ** 2
        vol_history.append(vol)
    
    # Fix: Ensure both arrays have the same length
    EWMA_df = pd.DataFrame({
        'returns': returns.values[1:], 
        'vol': np.sqrt(vol_history[1:len(returns)])  # Adjusted here
    })
    
    # Calculate VaR
    z_score = np.abs(np.percentile(returns, (1 - confidence_level) * 100))
    EWMA_df['VaR'] = -z_score * EWMA_df['vol']
    
    # Count exceptions
    EWMA_df['exception'] = EWMA_df['returns'] < EWMA_df['VaR']
    exception_count = EWMA_df['exception'].sum()
    total_obs = len(EWMA_df)
    expected_exceptions = total_obs * (1 - confidence_level)
    
    # Kupiec POF Test
    p = 1 - confidence_level
    x = exception_count
    n = total_obs
    LR_pof = -2 * np.log(((1 - p)**(n - x) * p**x) /
                         ((1 - (x / n))**(n - x) * (x / n)**x))
    p_value = 1 - chi2.cdf(LR_pof, df=1)
    
    # Results
    print(f"\nLambda: {lambda_value}")
    print(f"Number of exceptions: {exception_count}")
    print(f"Expected exceptions: {expected_exceptions}")
    print(f"Kupiec LR statistic: {LR_pof:.4f}")
    print(f"P-value: {p_value:.4f}")
    print("Model is acceptable" if p_value > 0.05 else "Model is NOT acceptable")
    
    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(EWMA_df['returns'].values, label='Returns')
    plt.plot(EWMA_df['VaR'].values, label='VaR (95%)', color='red')
    plt.scatter(EWMA_df.index[EWMA_df['exception']], EWMA_df['returns'][EWMA_df['exception']], 
                color='orange', label='Exceptions')
    plt.title(f'EWMA VaR Backtest (Lambda = {lambda_value})')
    plt.legend()
    plt.show()

    return EWMA_df

# Apply for lambda = 0.97
EWMA_df_97 = ewma_var_backtest(returns, lambda_value=0.97)

# Apply for lambda = 0.70
EWMA_df_70 = ewma_var_backtest(returns, lambda_value=0.70)

```

```python
import numpy as np
from scipy.stats import norm

# Parameters
lambda_value = 0.94
confidence_level = 0.95
z_score_95 = norm.ppf(confidence_level)  # Z-score for 95% confidence level
phi_95 = norm.pdf(z_score_95)            # PDF at Z-score for 95% confidence level

# Assuming daily_volatility is obtained from the EWMA model for 02/01/2025
daily_volatility = 0.02  # Replace with actual daily volatility from EWMA model

# 10-day volatility under the assumption of i.i.d. returns
vol_10_day = daily_volatility * np.sqrt(10)

# 10-day 95% VaR
VaR_10_day = -z_score_95 * vol_10_day

# 10-day 95% ETL (Expected Tail Loss)
ETL_10_day = -(phi_95 / (1 - confidence_level)) * vol_10_day

# Assuming max_volatility is the maximum observed daily volatility up to 02/01/2025
max_volatility = 0.03  # Replace with actual max daily volatility
MaxVaR_10_day = -z_score_95 * max_volatility * np.sqrt(10)

# Results
print(f"10-day 95% VaR: {VaR_10_day:.4f}")
print(f"10-day 95% ETL: {ETL_10_day:.4f}")
print(f"10-day 95% MaxVaR: {MaxVaR_10_day:.4f}")
```
