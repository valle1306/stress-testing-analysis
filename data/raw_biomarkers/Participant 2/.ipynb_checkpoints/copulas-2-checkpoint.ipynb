# plot points from a bivariate normal distribution

```python
def bivariate_normal(rho, n):
    """
    sample points from a bivariate normal distribution
    inputs: rhos=correlation, n=number of points
    """
    z1 = [normal() for i in range(n)]
    z2 = [normal() for i in range(n)]
    e1 = z1
    rho2 = math.sqrt(1 - pow(rho, 2))
    e2 = [rho * e1 + rho2 * e2 for e1, e2 in zip(z1, z2)]
    return e1, e2 

correlation = -.3
e1, e2 = bivariate_normal(correlation, 1000)
plt.scatter(e1,e2)
plt.title(f'correlation = {correlation:0.4}')
plt.show()
```

# define 2 univariate distributions

```python
class univariate():
    """univariate distribution, cumulative and inverse"""
    def __init__(self, fcn):
        self.a, self.b, self.steps = -3,3,100000  # number of steps needs to be very large otherwise may not cumsum to 1.0
        self.fcn = fcn
        self.x = np.linspace(self.a, self.b, self.steps)
        self.integral = pd.DataFrame([self.x, self.fcn(self.x)], index=['input','output']).transpose()
        self.integral['integral'] = np.cumsum(np.array(self.integral['output']))*(self.x[1]-self.x[0])

    def get_value(self, x):
        return self.fcn(x)
    
    def get_inverse(self, y):
        i = np.searchsorted(self.integral['integral'][:-1], y) ### exclude top most point  ##
        return float(self.integral['input'].iloc[i])

def plot_dist(g, function_name):
    x = np.linspace(-1, 2, 1000)
    out = pd.DataFrame([x, g(x)], index=['input','output']).transpose()
    fig, ax = plt.subplots(figsize=(18,5))
    plt.subplot(1,2,1)
    plt.plot(out['input'], out['output'], lw=5)
    plt.title(f'function {function_name} density')
    plt.subplot(1,2,2)
    plt.plot(x, np.cumsum(np.array(out['output']))*(x[1]-x[0]), lw=5)
    plt.title(f'function {function_name} distribution')
    plt.show()
    
def plot_2_dists(gs, function_names):
    plt.figure(figsize=(18,7))
    for i, (g, function_name) in enumerate(zip(gs, function_names)):
        x = np.linspace(-1, 2, 1000)
        out = pd.DataFrame([x, g(x)], index=['input','output']).transpose()
        plt.subplot(1,2,i+1)
        plt.plot(out['input'], out['output'], lw=5)
        plt.title(f'function {function_name} density', fontsize=15)
    plt.show()
```

```python
def f(x):
    return np.piecewise(x, 
                        [(-0.5 <= x) & (x <= 0.5),  # First range: -0.5 to 0.5
                         (0.5 < x) & (x <= 1.5),   # Second range: 0.5 to 1.5
                         (1.5 < x) & (x <= 1.751)],  # Third range: 1.5 to 1.751
                        [0.5, 0.25, 1.0])  # Corresponding values


def g(x):
    return np.piecewise(x, [x < 0, (x >= 0) & (x < 0.5), (x >= 0.5) & (x < 1.5), x > 1.5], 
                        [0, lambda x: x, lambda x: .5 - (x - 0.5) * 0.5, 0]) / .375


plot_dist(f, 'f1')
plot_dist(g, 'g1')
```

```python
plot_2_dists((f,g), ('f1','g1'))
```

```python
fcn_f = univariate(f)
fcn_g = univariate(g)
```

# create copula distribution (cor = -0.9)

```python
from scipy.stats import norm
from scipy.optimize import bisect

correlation = -.90
e1, e2 = bivariate_normal(correlation, 2000)
x1 = [fcn_f.get_inverse(norm.cdf(e)) for e in e1]
x2 = [fcn_g.get_inverse(norm.cdf(e)) for e in e2]

plt.figure(figsize=(18,4))
plt.subplot(1,2,1)
plt.hist(x1,30)
plt.title('x1 sample distribution')
plt.subplot(1,2,2)
plt.hist(x2,30)
plt.title('x2 sample distribution')
plt.show()

plt.figure(figsize=(7,7))
# plt.plot(x1,x2, 'o')
plt.scatter(x1, x2, s=80, facecolors='none', edgecolors='blue')
plt.title(f'copula with correlation {correlation}')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

# VaR

```python
def compute_var(x1, x2, correlation):
    """
    Computes the 95% Value at Risk (VaR) for the investment Z = X - 3Y.
    
    Parameters:
    - x1: Samples of X (first variable)
    - x2: Samples of Y (second variable)
    - correlation: The correlation used in the copula
    
    Returns:
    - 95% VaR value
    """
    Z = np.array(x1) - 3 * np.array(x2)  # Compute portfolio return
    var_95 = np.percentile(Z, 5)  # 5th percentile for 95% VaR
    mean_return = np.mean(Z)  # Compute the mean return
    
    print(f"Correlation {correlation}:")
    print(f"Mean Return: {mean_return:.4f}")
    print(f"95% VaR: {var_95:.4f}")
    print(f"Relative VaR (VaR - Mean): {var_95 - mean_return:.4f}")
    print("-" * 40)

# Compute VaR for correlation 0.9
compute_var(x1, x2, 0.9)

# Repeat the same process for correlation -0.9
correlation = -0.9
e1_neg, e2_neg = bivariate_normal(correlation, 2000)
x1_neg = [fcn_f.get_inverse(norm.cdf(e)) for e in e1_neg]
x2_neg = [fcn_g.get_inverse(norm.cdf(e)) for e in e2_neg]
compute_var(x1_neg, x2_neg, -0.9)

```

### 1. What is Vix index and what does it indicate about the current market?

* The Vix (cboe volatility index) is often referred to as the market's "fear gauge". It measures the expected volatility of the S&P 500 over the next 30 days, based on options pricing. A higher VIX indicates greater uncertainty and fear in the market, while a lower VIX suggests calm and stability

* In the article, VIX briefly spikes to 18.6 earlier this month but has since fallen to 15. The historical average is 19.5. This suggests that while there was some market anxiety, overall volatility remains relatively contained, which might indicate investor confidence despite the recent shock.

### 2. What is the "Magnificent seven"

* Meta, Nvidia, Microsoft, Apple, Amazon, Alphabet, Tesla

#### a. Why are their prices more volatile since the release of deepseek?

The volatility spike can be traced to DeepSeek's implications to the tech sector's fundamentals. The Magnificent seven had led the market in 2023, largely on optimism around AI. Companies like Nvidia were valued on the premise that AI development requires massive investments in chips and data centers- area these firms dominate. 

DeepSeek challenge that narrative by achieving AI breakthroughs with drastically lower cost and compute. This development raised question about the continued need for the massive data center capex spending by Big Tech. In other words, if AI can be done more efficiently, perhaps these companies won't see the runaway growth investors expected. 

The result was a gut punch to U.S. tech giants, as traders feared their expensive AI projects might not yield the competitive moat or profits once assumed. High valuations magnified the reaction: after huge 2023 run-ups, these stocks were priced for perfection, so any threat to their growth story caused outsized swings. Analysts at Invesco observed that the intensity of the sell-off reflected the massive expectationshigh valuations concentrated in U.S. tech. 

In sum, DeepSeeks debut introduced a jarring plot twist in the AI saga  one that ignited greater day-to-day price fluctuations as investors recalibrated their outlook for the Magnificent Seven

#### b. Why are some stocks such as Nvidia down in 2025?

Reversal after an AI-fueled boom: Several Magnificent Seven stocks have pulled back in 2025, and Nvidia is a prime example. In 2023, Nvidias shares soared on surging demand for its AI chips  the companys data-center revenue growth was parabolic
  making it one of the biggest winners of the AI boom. But by early 2025 the tide shifted. Every stock in the Magnificent Seven was down year-to-date (as of mid-February), even as the S&P 500 index was up about 24%. Nvidia, in particular, suffered a steep decline after the DeepSeek news, including the unprecedented one-day $600B wipeout mentioned earlier. There are a few key reasons for this reversal:

*DeepSeek undercut the AI scarcity narrative*: Nvidias valuation had assumed years of explosive growth selling high-end chips for AI. But DeepSeek showed that cutting-edge AI can be achieved with far fewer chips or alternative methods. Suddenly, investors questioned whether Nvidias future sales would be as robust as thought. The DeepSeek-driven sell-off was clearly about the sustainability of near-term growth in AI semiconductor manufacturers like Nvidia. In other words, if companies can do more with less hardware, the demand for Nvidias expensive GPUs might cool, hurting its growth prospects. This realization contributed heavily to Nvidias drop and the broader Magnificent Seven slump.

*Heavy spending vs. unclear payoff* : Even before DeepSeek, there were brewing concerns that Big Techs massive AI investments might not translate into proportional profits. The Magnificent Seven firms collectively plan to spend about $280 billion on AI and capex in 2025 (up 34% year-over-year), yet DeepSeek achieved a breakthrough with only a few million dollars in chips. This contrast has made Wall Street skeptical. As Bank of Americas strategists noted, recent price moves suggest growing concerns around monetization vs. capex for hyperscalers  i.e. investors fear these companies are pouring huge sums into AI without a guaranteed return. Nvidia embodies this issue: its customers (cloud giants like Amazon, Microsoft, Google) are spending heavily on AI infrastructure. If those customers start second-guessing big expenditures  a logical response to DeepSeeks efficiency  Nvidias sales could suffer. Thus, Nvidias stock decline in 2025 reflects a re-rating of its risk/reward, as investors weigh the possibility that AI spending (and Nvidias growth) might moderate sooner than expected.

*Rotation and profit-taking*: Another reason for the decline is simply rotation out of overextended tech names. After spectacular gains, many large investors have been locking in profits and shifting to other areas of the market. Goldman Sachs data showed hedge funds cutting back positions in the Magnificent Seven in early 2025 (Tesla was the lone exception), while increasing bets on sectors like healthcare and communications. Company insiders at the tech giants also started selling shares at the fastest pace since 2021. This exodus put additional downward pressure on stock prices. 

In Nvidias case, its extraordinary 2023 rally left it trading at high earnings multiples, which made it vulnerable. Once momentum broke  due to DeepSeeks shock and mixed early-2025 earnings  many traders took profits, accelerating the drop. In short, market leadership began to broaden out (toward cyclicals, value stocks, etc.), leaving last years AI darlings out of favor. This rotation has contributed to Nvidias slump and the Magnificent Sevens underperformance in 2025

Its worth noting that not every Magnificent Seven stock is down for the exact same reasons. For instance, Meta (Facebook) actually got a boost from strong earnings in early 2025, and Teslas issues may relate more to EV market dynamics. But broadly, the groups weakness stems from a comedown after 2024s euphoria. DeepSeek was the catalyst that exposed how stretched valuations and expectations had become. As one report put it, investors are still trying to appropriately quantify DeepSeeks impact on these companies  a process that has led to caution and lower stock prices. In summary, Nvidia and its peers have declined in 2025 due to a mix of AI competition fears, skepticism about huge capital spending, and a natural rotation away from overcrowded trades.

#### c.  Why is this increase in volatility not transmitted to the overall index?

* The sell-off was narrow, not systemic: The DeepSeek shock, while intense, was focused mainly on big tech. Many other stocks and sectors were unaffected  some even benefited as investors rotated money elsewhere. Notably, on the day Nvidia collapsed in late January, roughly 70% of S&P 500 stocks actually rose (including three of the Magnificent Seven stocks themselves). In other words, a majority of the market was divorced from the tech plunge, so the indexs overall move was much smaller than the headline-grabbing drops in Apple, Nvidia, etc. As Invescos strategists observed, the one-day sell-off was narrow, which kept it from sparking a broader panic. This dispersion of outcomes  some stocks down big, many others up  diluted the volatility when you look at the entire index.

* Broad market strength offset tech weakness: In early 2025 the U.S. economy showed solid momentum (helped by prior Fed rate cuts and fiscal stimulus), which boosted cyclicals, industrials, travel companies, and other non-tech shares. At the same time, bond yields eased off their highs, which helped interest-rate-sensitive sectors. As a result, a wider variety of stocks within the S&P 500 gaining, even as the Magnificent 7 have struggled, according to Morgan Stanley. This improved breadth meant the index could keep rising (or at least hold steady) despite losses in the top-weighted tech names. By mid-February 2025, the S&P 500 was up roughly 4% year-to-date with all 11 sectors in the green  a remarkably resilient performance considering the hit that mega-cap tech took. Simply put, other stocks picked up the slack. For example, healthcare and communication services stocks, which lagged in 2024, saw strong gains in early 2025, and momentum favorites like Palantir surged to new highs. This rotation into other winners buffered the S&P 500, making the overall index far less volatile than the tech subset.

* Index math and weighting effects: Its also important to remember how the S&P 500 is constructed. The Magnificent Seven are large  together they accounted for about one-third of the indexs market cap at the start of 2025. However, even at 33%, they are not 100%. The other ~493 companies collectively make up the remaining two-thirds. When the Magnificent Seven dropped sharply, that did drag the index down, but only in proportion to their weight. For instance, Nvidias crash alone trimmed the S&P, but could not outweigh gains in hundreds of smaller names. Moreover, the Magnificent Seven didnt all move in lockstep; on some days a couple of them rose (or fell less) while others plunged. This lack of perfect correlation further reduced the volatility transmission. In effect, the S&Ps diversification blunted the impact. As one market strategist noted, the index traded in a relatively narrow range in early 2025  roughly back to its December levels  even amid the tech turmoil. Diversification and rotation ensured that the overall markets volatility stayed moderate, even though a few superstar stocks were on a roller coaster.


### **3. What was the "meme stock" hangover?**

The **meme stock hangover** refers to the period following the **meme stock frenzy of 2021**, during which heavily shorted and retail-favored stocks, such as **GameStop (GME)** and **AMC Entertainment (AMC)**, experienced massive volatility due to coordinated buying by retail investors on platforms like **Reddit's WallStreetBets**. The hangover phase occurred **in 2022**, when these stocks **collapsed from their peak prices**, leading to significant losses for latecomers to the rally and shifting sentiment among retail investors.

---

#### a. Which stock or stocks were most affected during that period in 2022?
The most affected stocks during the **meme stock hangover in 2022** were:

1. **GameStop (GME)**  The original meme stock, which surged to over **$400 per share in early 2021**, saw a sharp decline in 2022 as enthusiasm faded.  
2. **AMC Entertainment (AMC)**  AMC, which skyrocketed due to retail buying and a short squeeze in 2021, suffered major losses as its fundamentals remained weak and dilution concerns increased.  
3. **Bed Bath & Beyond (BBBY)**  Another heavily shorted stock that gained meme status but collapsed in 2022 due to **poor financial health** and liquidity concerns.  
4. **BlackBerry (BB)** and **Koss (KOSS)**  These stocks briefly saw gains in the 2021 meme craze but followed the broader trend of decline in 2022.  

During the hangover, these stocks experienced **severe drawdowns**, as the initial retail-driven momentum faded and reality set in regarding the companies weak fundamentals. **For example, GME and AMC lost over 70% of their peak values in 2022.**  

---

#### b. Did the dispersion index go up or down during that time period? Why?
The **dispersion index** measures how much individual stocks move relative to the broader market. 

During the **meme stock surge of 2021**, **dispersion spiked**, as a small set of stocks (like GME and AMC) made extreme moves independent of overall market trends. However, **in 2022, during the meme stock hangover, the dispersion index declined.**  

**Why did dispersion go down?**
1. **Meme stocks reverted to broader market trends**  As speculative enthusiasm faded, these stocks began moving more in line with general market forces, reducing their outsized impact on dispersion.  
2. **Retail trading activity declined**  Many retail traders suffered losses, leading to less coordination and fewer parabolic moves, which lowered dispersion.  
3. **Market-wide risk factors dominated**  The focus in 2022 shifted toward macroeconomic concerns, such as **rising interest rates, inflation, and recession fears**, which caused a **more uniform sell-off** rather than isolated volatility in meme stocks.  

Thus, while **dispersion spiked in 2021**, it **declined in 2022**, as the meme stock effect faded and macroeconomic conditions dictated broader market moves.

### 4.

To **detect irrational exuberance** in the stock market, we need a measure that captures **excessive speculation, unsustainable price surges, and deviation from fundamentals**. A strong candidate for such a measure is the **Retail Sentiment & Bubble Indicator (RSBI)**, which combines multiple quantitative metrics to assess speculative behavior.  

### **Proposed Measure: Retail Sentiment & Bubble Indicator (RSBI)**
The RSBI would be constructed from the following components:  

1. **Options Volume Skew (OVS):**  
   - Measures the ratio of call option volume to put option volume in speculative stocks.  
   - A **sharp increase in call option buying** (especially in out-of-the-money contracts) signals **excessive bullish speculation**, often driven by retail traders.  
   - Example: In early 2021, GameStop (GME) saw call option volume **10x higher than historical averages**, fueling its short squeeze.  

2. **Retail Trading Flow Index (RTFI):**  
   - Tracks the **proportion of total market volume** coming from **retail brokerage platforms** (e.g., Robinhood, Webull).  
   - A sudden **spike in retail participation** (beyond historical norms) suggests irrational exuberance.  
   - Example: During the meme stock frenzy, retail traders accounted for **40%+ of total market volume**, up from ~20% in normal periods.  

3. **Social Media Hype Score (SMHS):**  
   - Uses natural language processing (NLP) on **Reddit (WallStreetBets), Twitter, StockTwits**, and other platforms to quantify **hype intensity**.  
   - Key indicators include **explosive growth in mentions, sentiment score shifts, and engagement surges**.  
   - Example: GMEs daily mentions **skyrocketed from 200 to 500,000+ in January 2021** just before its parabolic rise.  

4. **Price-Fundamental Divergence (PFD):**  
   - Measures the deviation of a stocks price from its **intrinsic valuation** (based on earnings, cash flow, and analyst estimates).  
   - A stock trading **5x or more above its historical valuation multiples** signals **speculative excess**.  
   - Example: AMC traded at **100x+ forward earnings in mid-2021**, far above its historical range (~10-15x).  

5. **Short Interest & Gamma Squeeze Risk Index (SIGSRI):**  
   - Tracks stocks with **extremely high short interest (above 50% of float)** and **heavy call option positioning**a key setup for short squeezes.  
   - A spike in this index indicates **potential for rapid, irrational price movements**, detached from fundamentals.  
   - Example: GameStops short interest was **140% of float** in early 2021, leading to its historic short squeeze.  

### **Implementation & Thresholds:**
- Compute each component daily and normalize scores between **0 and 100**.  
- A **RSBI above 80** indicates **highly speculative market conditions**, likely driven by **irrational exuberance**.  
- Backtests would confirm **historical correlations** between RSBI spikes and market bubbles.  

### **Why RSBI Works?**
Unlike traditional valuation metrics (like P/E ratios), **RSBI directly measures speculative excess** by analyzing **retail activity, options positioning, sentiment, and price-fundamental dislocations**. It would have successfully flagged past speculative frenzies**dot-com bubble (1999-2000), meme stock mania (2021), and crypto surges (2021, 2023).**


### 5.

### **Investment Strategy in a High Dispersion Environment**
In periods of **high dispersion**, individual stock returns **vary widely**, meaning some stocks **significantly outperform or underperform** the broader market. Given this dynamic, investors should focus on **stock selection rather than passive index investing**.

### **Why Choose Specific Stocks Over an Index ETF?**
1. **Greater Opportunity for Outperformance**  
   - In high-dispersion markets, the performance gap between the best and worst stocks **widens**.  
   - Skilled investors can **capitalize on mispricings** by selecting strong performers while avoiding weak ones.  
   - Example: During the **tech bubble burst (2000-2002)** and the **COVID market recovery (2020-2021)**, stock dispersion was high, allowing active stock pickers to beat the index.

2. **Index ETFs Become Less Attractive**  
   - If stock returns **diverge significantly**, the overall index **averages out the winners and losers**, leading to **muted returns**.  
   - Example: In early 2025, the **S&P 500 remained flat despite major swings in individual stocks** because losses in tech stocks were offset by gains in cyclicals.  
   - Investors who bought an **S&P 500 ETF (e.g., SPY, VOO)** would have experienced **lower returns** than those who selectively invested in outperforming stocks.

3. **Sector Rotation & Thematic Investing**  
   - High dispersion often occurs during **market transitions**, where certain sectors **outperform others significantly**.  
   - Investors can **adjust portfolios** to favor **emerging winners** while reducing exposure to declining industries.  
   - Example: **In 2022, energy stocks surged while tech stocks plummeted**index ETFs that held both sectors delivered mediocre returns, but sector-specific plays (like XLE for energy) thrived.

4. **Hedge Fund & Institutional Strategies in High Dispersion Markets**  
   - Hedge funds often **increase their active bets** during high dispersion, as the market presents more **mispriced opportunities**.  
   - **Market-neutral or long-short strategies** benefit in this environment by **going long on strong stocks and shorting weak ones**.

### **When Would an Index ETF Still Make Sense?**
- If an investor **lacks the expertise or time** to analyze individual stocks.
- If they prefer **low-cost, diversified exposure** rather than taking on stock-specific risk.
- If dispersion is high **but random** (i.e., no clear patterns for sector or stock outperformance).

### **Conclusion: Stock Picking Wins in High Dispersion Markets**
In high dispersion periods, **active stock selection offers higher return potential** than passive index ETFs. Investors who can **identify trends, sector shifts, and high-quality stocks** stand to benefit **significantly more** than those who simply hold a broad-market ETF. However, for those seeking lower risk and **broad diversification**, **index ETFs remain a safer choice**.
